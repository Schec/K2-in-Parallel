```python
import numpy as np
import math
import time
import random
import collections
import itertools
import pandas as pd
from __future__ import division
```

```python
#PARAMETERS

#Data
D = np.random.randint(2, size=(100,10))
D = pd.DataFrame(D)
n = D.shape[1]
u = n-1

#Create Automatically Pred
Pred = []
r = []
for i in range(n):
    Pred.append(range(i))
    r.append(len(set(D[i])))

pi = {}
for i in xrange(n):
    pi[str(i)] = []

#K2 - algorithm
for i in xrange(n):
    #print '-------------------'+str(i)
    
    #Initialization of P_old
    unique_vals = D[i].unique()
    alphas = []
    for ele in unique_vals:
        alphas.append(sum([x==ele for x in D[i]]))
    N = sum(alphas)#J'aurais pu le définir de pmanière plus simple...
    P_old = math.factorial(r[i]-1)/math.factorial(N+r[i]-1)*np.product([math.factorial(x) for x in alphas])
    #print 'P_old = %f' % P_old
    
    
    #START
    OKToProceed = True
    the_list = []
    while OKToProceed and len(pi[str(i)])<u and len(Pred[i])>0:
        vals = []
        Dyn_Pred = [x for x in Pred[i] if x not in pi[str(i)]]
        #print 'Dyn_Pred for %i: %s'%(i,str(Dyn_Pred))
        for pred in Dyn_Pred:
            unique = D[pred].unique()
            temp = [x for x in the_list]
            temp.append(unique)
            combinations = list(itertools.product(*temp))
            val1 = 1
            temp_nodes = [x for x in pi[str(i)]]
            temp_nodes.append(pred)
            for ele in combinations:
                temp_D = D
                #print '----'+str(pred)
                #print ele
                for l in range(len(ele)):
                    temp_D = temp_D[temp_D[temp_nodes[l]]==ele[l]]
                #print i
                #print temp_D
                SS = []
                for uval in unique_vals:
                    SS.append(sum(temp_D[i]==uval))
                NN = sum(SS)
                prod_fact = np.prod([math.factorial(x) for x in SS])
                val1 *= math.factorial(r[i]-1)/math.factorial(NN+r[i]-1)*prod_fact
            vals.append(val1)
        if len(vals)>0:
            P_new = np.max(vals)
            p = [ww for ww, j in enumerate(vals) if j == P_new]
            pos = p[0]
            if P_new > P_old:
                P_old = P_new
                pi[str(i)].append(Dyn_Pred[pos])
                the_list.append(list(set(D[pred])))
                if len(pi[str(i)])==len(Pred[i]):
                          OKToProceed = False
            else:
                OKToProceed = False
        else:
            OKToProceed = False
    
    #if len(pi[str(i)])>0:
        #for ele in pi[str(i)]:
            #print 'Node: '+str(i)+', Parents of '+str(i)+': '+str(ele)
    #else:
        #print 'Node: '+str(i)+' has no parents!!'
      
```
